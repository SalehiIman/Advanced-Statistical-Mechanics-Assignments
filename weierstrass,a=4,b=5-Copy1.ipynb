```python
from time import time
t1=time()
from numpy import *
import random as r
from math import *
%matplotlib inline
import matplotlib.pyplot as plt
N=1000;n=5;a=4;b=5;delta=1;m=3000000
p_n=[((a-1)/float(a**(j+1))) for j in range(n+1)]
P_n=[sum(p_n[0:i]) for i in arange(1,n+1)] b
X=[]
for q in range(m):
    l=[]
    P_l=[r.random() for i in range(int(N))]
    for j in range((N)):
        for g in range(n):
            if P_l[j]<=P_n[g]:
                l.append((b**(g))*delta)
                break
        if P_l[j]>P_n[-1]:
            l.append((b**(n))*delta)
    x=0
    for s in range(len(l)):
        if r.random()>=.5:
            x+=l[s]
        else:
            x-=l[s]
    X.append(x)
fig=plt.figure()
fig,axes1=plt.subplots(figsize=(12,6))
fig.suptitle('Weierstrass One_Dim Random Walk_ N={}'.format(m))
axes1.hist(X,bins=100,normed=True,color='red')
axes1.set_xlabel('X')
axes1.set_ylabel('PDF')
axes1.legend(['$a=4,b=5$\n$a\leq b^{2}\Rightarrow Levy$ $ Distribution$\n $0<\mu =Ln(a)/Ln(b)<1$\n       \
   $\Downarrow$\n $fractal$ $like$ $structure$'])
plt.show()
t2=time()-t1
t2=time()-t1
print('T:'),t2
```
