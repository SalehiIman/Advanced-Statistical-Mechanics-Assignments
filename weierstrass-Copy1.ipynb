```python
from numpy import *
import random as r
from math import *
%matplotlib inline
import matplotlib.pyplot as plt
n=5;a=4;b=5;delta=1;m=100#number of walkers
p_n=[((a-1)/float(a**(j+1))) for j in range(n+1)]
P_n=[sum(p_n[0:i]) for i in arange(1,n+1)]
T_counter=[]
fig,ax=plt.subplots()
for i in range(m):
    X=[0]
    c=0;x=0;t_counter=0;s=0
    while c==0:
        p_l=r.random()
        for g in range(n):
            if p_l<=P_n[g]:
                l=((b**(g))*delta)
                break
            if p_l>P_n[-1]:
                l=((b**(n))*delta)
        if r.random()>=.5:
            X.append(X[s]+l)
            x+=l
        else:
            X.append(X[s]-l)
            x-=l
        s+=1
        if x==0 and s>1:
            c+=1
        t_counter+=1
    T_counter.append(t_counter)
    ax.plot(X)
fig.suptitle('one_dim random walk with condition of riturning to point zero_{}walkers'.format(m))
ax.set_xlabel('time')
ax.set_ylabel('X')
plt.show()
T_Avg=sum(T_counter)/float(len(T_counter))
print('Avg Time for Returning to  Point Zero for {}walkers:'.format(m)),T_Avg
```
